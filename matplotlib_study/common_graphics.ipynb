# matplotlib画图

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

创建DataFrame

```python
data = pd.DataFrame({'time':pd.date_range('20170101','20170110'),
                     'value':np.random.randn(10)})
```

## 折线图

```python
plt.plot(data['time'],data['value'])#第一个参数是横轴，第二个参数是纵轴，横轴可省略
plt.show()
```

### 画图细节（增加横纵坐标名称、图表名称以及坐标轴标签旋转）

```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['KaiTi']#作图的中文
matplotlib.rcParams['font.serif'] = ['KaiTi']#作图的中文
plt.rcParams['axes.unicode_minus'] = False#正常显示负号
plt.plot(data['time'],data['value'])
plt.xticks(rotation = 90)#坐标轴标签旋转
plt.xlabel('日期')#横坐标名称
plt.ylabel('下雨概率')#纵坐标名称
plt.title('天气预报')#图表名称
plt.show()
```

## 多图并列

```python
fig = plt.figure(figsize = (8,4))#先建立画布,大小为8*4
ax1 = fig.add_subplot(2,2,1)#在第一个区域建立轴
ax2 = fig.add_subplot(2,2,2)#在第二个区域建立轴
ax3 = fig.add_subplot(2,2,4)#在第四个区域建立轴
ax1.plot(np.random.randn(6))
ax2.plot(np.random.randn(6))
ax3.plot(np.random.randn(6))
plt.show()
```

## 多图重合、颜色设定、类别标签

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
date_time = pd.date_range('20170101','20170110')
for i in range(5):
    label = str(201801 + i)
    plt.plot(date_time, np.random.randn(10), c=colors[i], label=label)
plt.legend(loc='best')
plt.xlabel('日期')
plt.ylabel('降雨量')
plt.title('各月份降雨量, 1948-1952')

plt.show()
```

```python
data2 = pd.DataFrame(np.random.randn(120).reshape(20,6),columns=list('ABCDEF'))
```

## 柱形图

```python
fig, ax=plt.subplots()
bar_heights = data2.sum().values+10#每根柱子的高度，也是数据主体
bar_position=range(1,7)#每根柱子在X轴的位置
ax.set_xticks(bar_position)#每个x轴标签的位置
ax.set_xticklabels(data2.columns)#每个x轴标签
ax.bar(bar_position,bar_heights,0.8)#画图，第三个参数是柱子宽度
plt.show()
```

## 散点图

```python
fig, ax = plt.subplots()
ax.scatter(np.random.randn(100), np.random.randn(100))#每个点的X轴和Y轴
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
plt.show()
```

## 直方图

```python
fig, ax = plt.subplots()
ax.hist(np.random.randn(1000))#直方图
#ax.hist(np.random.randn(1000),range=(0,1),bins=20) #range代表只取范围内的，bins是柱子的个数
plt.show()
```

## 箱线图

```python
fig, ax = plt.subplots()
ax.boxplot(data2.values)
ax.set_xticklabels(data2.columns)#增加横坐标标签
plt.show()
```
