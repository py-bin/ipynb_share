```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('测试数据.xlsx')
data.head()
```

```python
data.info()
```

```python
data['日期']=data['日期'].astype(str)
```

```python
data['订单月份']=data['日期'].str.slice(0,6)
data['订单月份'].head()
```

```python
data['地市']=data['地市'].str.slice(0,2)
data['地市'].head()
```

```python
data2 = data['来源'].str.split('-',expand = True)
data2.head()
```

```python
data2.columns = ['场景','业务类型']
data2.head()
```

```python
data = pd.concat([data,data2],axis =1)
data.head()
```

```python
data['服务内容']=' '+data['服务内容']+' '
```

```python
data['电话号码']=data['服务内容'].str.findall(r'\D(1\d{10})\D')
data['电话号码']
```

```python
data['电话号码']=data['电话号码'].str.join('/')
```
